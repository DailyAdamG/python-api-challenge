```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
from scipy.stats import linregress

from api_keys import weather_api_key

from citipy import citipy

#Output file

output_data_file = "../output_data/cities.csv"

#Range of values

lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
#Lists for holding values

lat_lngs = []
cities = []

#Random city generator

lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

#Finding nearest city for each coordinate combination

for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    if city not in cities:
        cities.append(city)
                  
#Print city count
len(cities)
```

```python
#Place to store the data from for loop

count = 0
new_cities = []
latitude = []
longitude = []
max_temp = []
humidity = []
cloudiness = []
wind_speed = []
country = []
date = []

base_url = 'http://api.openweathermap.org/data/2.5/weather?'
units = "imperial"

#Loop to grab the information from the API

for city in cities:
    url = f"{base_url}appid={weather_api_key}&units={units}&q={city}"
    response = requests.get(url).json()

#Checking to ensure cities have weather data, omitting cities that do not have weather data

    try:
        latitude.append(response["coord"]["lat"])
        longitude.append(response["coord"]["lon"])
        max_temp.append(response["main"]["temp_max"])
        humidity.append(response["main"]["humidity"])
        cloudiness.append(response["clouds"]["all"])
        wind_speed.append(response["wind"]["speed"])
        country.append(response["sys"]["country"])
        date.append(response["dt"])
        new_cities.append(city)
        count = count + 1
        print(f"Processing city {count}: {city}")
        
    except KeyError:
        print(f"City not found, skipping {city}")
```

```python
#Creating a dataframe


weather_df = pd.DataFrame({"City" : new_cities,
                         "Latitude" : latitude,
                         "Longitude" : longitude,
                         "Max Temp (Fahrenheit)" : max_temp,
                         "Humidity" : humidity,
                         "Cloudiness" : cloudiness,
                          "Wind Speed (MPH)" : wind_speed,
                           "Country" : country,
                           "Date" : date})

#Double-checking the dataframe for missing values

weather_df.count()

weather_df
```

```python
# Checking for humidity over 100

cities_over_humidity = weather_df["City"].loc[weather_df["Humidity"] > 100]

if cities_over_humidity.count() > 0:
    print(f"There are {cities_over_humidity.count()} cities over 100% humidity. We have thrown them out.")
    new_weather_df = weather_df.loc[weather_df["Humidity"] < 100]
    print(new_weather_df)

else: print(f"There are no cities with over 100% humidity.")
```

```python
#Exporting the data to csv file

weather_df.to_csv("output_data/cities.csv")
```

```python
#Splitting the data into Northern and Southern Hemispheres

northern_df = weather_df.loc[weather_df["Latitude"] > 0]
southern_df = weather_df.loc[weather_df["Latitude"] < 0]
```

```python
#Creating variable for regression

slope, y_int, r_value, p_value, std_err = linregress(northern_df["Latitude"], northern_df["Max Temp (Fahrenheit)"])

north_temp_fit = slope * northern_df["Latitude"] + y_int

#Regression equation

north_temp_equation = "y= " + str(round(slope,2)) + "x + " + str(round(y_int,2))

#Getting R-squared value

r_square = "R-Squared Value is " + str(abs(round(r_value,2)))
print(r_square)

#Creating the scatter plot

plt.scatter(northern_df["Latitude"], northern_df["Max Temp (Fahrenheit)"], edgecolors = "black", facecolor = "red")
plt.plot(northern_df["Latitude"], north_temp_fit, color = "blue", linewidth = 3)
plt.annotate(north_temp_equation, (-8, 40), color = "blue", fontsize = 16)
plt.title("City Latitude vs. Max Temperature (Northern Hemisphere)")
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (F)")
plt.xlim(-10,90)
plt.savefig("output_data/Northern Lat vs. Temp.png")
plt.show()

#This plot is showing that there is a strong negative correlation between latitude and max temperature.
#As you go further North, the lower the maximum temperature.
```

```python
#Creating variable for regression

slope, y_int, r_value, p_value, std_err = linregress(southern_df["Latitude"], southern_df["Max Temp (Fahrenheit)"])

south_temp_fit = slope * southern_df["Latitude"] + y_int

#Regression equation

south_temp_equation = "y= " + str(round(slope,2)) + "x + " + str(round(y_int,2))

#Getting R-squared value

r_square = "R-Squared Value is " + str(abs(round(r_value,2)))
print(r_square)

#Creating the scatter plot

plt.scatter(southern_df["Latitude"], southern_df["Max Temp (Fahrenheit)"], edgecolors = "black", facecolor = "red")
plt.plot(southern_df["Latitude"], south_temp_fit, color = "blue", linewidth = 3)
plt.annotate(south_temp_equation, (-62, 72), color = "blue", fontsize = 16)
plt.title("City Latitude vs. Max Temperature (Southern Hemisphere)")
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (F)")
plt.xlim(-65,5)
plt.savefig("output_data/Southern Lat vs. Temp.png")
plt.show()

#This plot is showing that there is a decent positive correlation between latitude and max temperature.
#As you go further South, the lower the maximum temperature.
```
